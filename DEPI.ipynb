```python
for row in range(10):
    for column in range(10):
        if (column == 0 or column == 6) and row != 0 or (row == 0 or row == 5) and (column > 0 and column < 6):
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
for row in range(10):
    for column in range(10):
        if ((column == 2 or column == 6) and row > 2) or (row == 2 or row == 5) and (column > 2 and column < 6):
            print(" ", end=" ")
        else:
            print("*", end=" ")
    print()
```

```python
low, high = 50, 80
for x in range(low, high + 1):
    if x > 1:
        for i in range(2, x):
            if (x % i) == 0:
                break
        else:
            print(x)
```
